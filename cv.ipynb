```python
import cv2 as cv
import numpy as np
import os
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow
from tensorflow.keras import models, layers
```

```python
adidas_train = [f"archive/train/adidas/{i}" for i in os.listdir("archive/train/adidas")]
nike_train = [f"archive/train/nike/{i}" for i in os.listdir("archive/train/nike")]
converse_train = [f"archive/train/converse/{i}" for i in os.listdir("archive/train/converse")]
```

```python
img_dict_train = {
    "adidas": adidas_train,
    "nike": nike_train,
    "converse": converse_train
}

img_train_df = pd.DataFrame(img_dict_train)
```

```python
img_train_df
```

```python
rows = len(img_train_df.iloc[:, 0])
```

```python
classes = img_train_df.columns
```

```python
for i in range(len(classes)):
    j = 0
    plt.imshow(cv.imread(img_train_df[classes[i]][j]))
    # add class in title
    plt.title(classes[i])
    plt.show()
```

```python
img_dict_test = {
    "adidas": [f"archive/test/adidas/{i}" for i in os.listdir("archive/test/adidas")],
    "nike": [f"archive/test/nike/{i}" for i in os.listdir("archive/test/nike")],
    "converse": [f"archive/test/converse/{i}" for i in os.listdir("archive/test/converse")]
}

img_test_df = pd.DataFrame(img_dict_test)

img_test_df
```

```python
# add this data for CNN
train_data = []
train_labels = []

for i in range(len(classes)):
    for j in range(len(img_train_df[classes[i]])):
        train_data.append(cv.imread(img_train_df[classes[i]][j]))
        train_labels.append(i)

train_data = np.array(train_data)
train_labels = np.array(train_labels)

print(train_data.shape)
print(train_labels.shape)

test_data = []
test_labels = []

for i in range(len(classes)):
    for j in range(len(img_test_df[classes[i]])):
        test_data.append(cv.imread(img_test_df[classes[i]][j]))
        test_labels.append(i)

test_data = np.array(test_data)
test_labels = np.array(test_labels)

print(test_data.shape)

print(test_labels.shape)
```

```python
# normalize data
train_data = train_data / 255.0
test_data = test_data / 255.0

# build model
model = models.Sequential([
    layers.Conv2D(filters=32, kernel_size=(3, 3), activation='relu', input_shape=(240, 240, 3)),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(filters=64, kernel_size=(3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(filters=128, kernel_size=(3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dense(3, activation='softmax')
])

model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

model.summary()

# train model
history = model.fit(train_data, train_labels, epochs=15, validation_data=(test_data, test_labels))
```

```python
# plot accuracy and loss
plt.plot(history.history['accuracy'], label='accuracy')
plt.plot(history.history['val_accuracy'], label='val_accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.ylim([0, 1])
plt.legend(loc='lower right')
plt.show()
```

```python
# make a bounding box around the object
def find_class(shoe, num):
    img = cv.imread(f'archive/test/{shoe}/{num}.jpg')
    plt.imshow(img)
    plt.show()

    img = cv.resize(img, (240, 240))
    img = np.reshape(img, (1, 240, 240, 3))
    img = img / 255.0

    prediction = model.predict(img)
    print(prediction)

    print(np.argmax(prediction))

    # get the index of the class with the highest probability score

    class_index = np.argmax(prediction)

    # get the probability score of the predicted class

    probability_score = np.max(prediction)

    # get the name of the predicted class

    predicted_class = classes[class_index]

    # print the class name and probability score

    print("This is a {} with {:.2f} percent confidence.".format(predicted_class, probability_score * 100))
```
